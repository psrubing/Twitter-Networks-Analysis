```python
import os
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn import svm
import pickle
from nltk.corpus import stopwords
```

```python
# prepare dataset
```

```python
anti_accounts_path = r'C:\Users\psrub\Documents\Python\Twitter\anti_scientific_data_2'
hand_picked_new_anti_hubs_path = r'C:\Users\psrub\Documents\Python\Twitter\hand_picked_new_anti_hubs'

pro_accounts_path = r'C:\Users\psrub\Documents\Python\Twitter\pro_scientific_data_2'
hand_picked_new_pro_hubs_path = r'C:\Users\psrub\Documents\Python\Twitter\hand_picked_new_pro_hubs'

neutral_accounts_path = r'C:\Users\psrub\Documents\Python\Twitter\neutral_data_1'
```

```python
anti_accounts = os.listdir(anti_accounts_path)
hand_picked_new_anti_hubs = os.listdir(hand_picked_new_anti_hubs_path)

pro_accounts = os.listdir(pro_accounts_path)
hand_picked_new_pro_hubs = os.listdir(hand_picked_new_pro_hubs_path)

neutral_accounts = os.listdir(neutral_accounts_path)
```

```python
# ANTI BASE
tweets_dfs = []
for account in anti_accounts:
    acc_path = os.path.join(anti_accounts_path, account)
    try: 
        print(os.path.join(acc_path, 'tweets.json'))
        df = pd.read_json(os.path.join(acc_path, 'tweets.json'))
        df['account'] = account
        tweets_dfs.append(df)
    except ValueError:
        continue

# ANTI NEW
for account in hand_picked_new_anti_hubs:
    acc_path = os.path.join(hand_picked_new_anti_hubs_path, account)
    try: 
        print(os.path.join(acc_path, 'tweets.json'))
        df = pd.read_json(os.path.join(acc_path, 'tweets.json'))
        df['account'] = account
        tweets_dfs.append(df)
    except ValueError:
        continue

anti_df = pd.concat(tweets_dfs, ignore_index=True)
anti_df['class'] = 0


# PRO BASE
tweets_dfs = []
for account in pro_accounts:
    acc_path = os.path.join(pro_accounts_path, account)
    try: 
        print(os.path.join(acc_path, 'tweets.json'))
        df = pd.read_json(os.path.join(acc_path, 'tweets.json'))
        df['account'] = account
        tweets_dfs.append(df)
    except ValueError:
        continue

# PRO NEW
for account in hand_picked_new_pro_hubs:
    acc_path = os.path.join(hand_picked_new_pro_hubs_path, account)
    try: 
        print(os.path.join(acc_path, 'tweets.json'))
        df = pd.read_json(os.path.join(acc_path, 'tweets.json'))
        df['account'] = account
        tweets_dfs.append(df)
    except ValueError:
        continue
        
pro_df = pd.concat(tweets_dfs, ignore_index=True)
pro_df['class'] = 2


tweets_dfs = []
for account in neutral_accounts:
    acc_path = os.path.join(neutral_accounts_path, account)
    try: 
        print(os.path.join(acc_path, 'tweets.json'))
        df = pd.read_json(os.path.join(acc_path, 'tweets.json'))
        df['account'] = account
        tweets_dfs.append(df)
    except ValueError:
        continue
        
neutral_df = pd.concat(tweets_dfs, ignore_index=True)
neutral_df['class'] = 1

df = pd.concat([anti_df, pro_df, neutral_df], ignore_index=True)
```

```python
df
```

```python
# FILTER TWEETS FROM STOPWORDS
```

```python
english_stopwords = stopwords.words("english")
```

```python
df['filtered_text'] = df['clean_text'].apply(lambda x: ' '.join([word for word in x.split() if word not in english_stopwords]))
```

```python
df
```

```python
X_train, X_test, y_train, y_test = train_test_split(df['clean_text'], df['class'], test_size=0.15, random_state=42)
```

```python
vectorizer = TfidfVectorizer(use_idf=True)
```

```python
tfidf_train = vectorizer.fit_transform(X_train)
tfidf_test = vectorizer.transform(X_test)
```

```python
# SVM

svmc = svm.SVC(probability=True)
svmc.fit(tfidf_train, y_train)
svmc_y_predict = svmc.predict(tfidf_test)
svmc_y_predict_proba = svmc.predict_proba(tfidf_test)

print(classification_report(y_test, svmc_y_predict))
print(confusion_matrix(y_test, svmc_y_predict))
```

```python
pickle.dump(svmc, open('svm_all_tweets.h5', 'wb'))
```

```python
# RandomForest

rfc = RandomForestClassifier()
rfc.fit(tfidf_train, y_train)
rfc_y_predict = rfc.predict(tfidf_test)
rfc_y_predict_proba = rfc.predict_proba(tfidf_test)

print(classification_report(y_test, rfc_y_predict))
print(confusion_matrix(y_test, rfc_y_predict))
```

```python
pickle.dump(rfc, open('rfc_all_tweets.h5', 'wb'))
```

```python
# Merge text per account
```

```python
df_grouped_text = df.groupby(['account', 'class'], as_index = False).agg({'filtered_text': ' '.join})
```

```python
df_grouped_text
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_grouped_text['filtered_text'], df_grouped_text['class'], test_size=0.15, random_state=42)
```

```python
vectorizer = TfidfVectorizer(use_idf=True)
```

```python
tfidf_train = vectorizer.fit_transform(X_train)
tfidf_test = vectorizer.transform(X_test)
```

```python
# SVM on merged tweets with 3 classes and new pro and anti hubs accounts from hand picking

svmc = svm.SVC(probability=True)
svmc.fit(tfidf_train, y_train)
svmc_y_predict = svmc.predict(tfidf_test)
svmc_y_predict_proba = svmc.predict_proba(tfidf_test)

print(classification_report(y_test, svmc_y_predict))
print(confusion_matrix(y_test, svmc_y_predict))
```

```python
pickle.dump(svmc, open('svm_merged_tweets_3_classes_with_new_hubs.h5', 'wb'))
```

```python
# CROSSVALIDATION
```

```python
skf = StratifiedKFold(n_splits=5, random_state=1, shuffle=True)

y = df_grouped_text['class']

for train_index, test_index in skf.split(df_grouped_text['filtered_text'], y):
    
    train = df_grouped_text.loc[train_index]
    test = df_grouped_text.loc[test_index]
    
    vectorizer = TfidfVectorizer(use_idf=True)
    
    tfidf_train = vectorizer.fit_transform(train['filtered_text'])
    tfidf_test = vectorizer.transform(test['filtered_text'])
    
    svmc = svm.SVC(probability=True)
    svmc.fit(tfidf_train, y.loc[train_index])
    svmc_y_predict = svmc.predict(tfidf_test)
    svmc_y_predict_proba = svmc.predict_proba(tfidf_test)
    
    print(classification_report(y.loc[test_index], svmc_y_predict))
    print(confusion_matrix(y.loc[test_index], svmc_y_predict))
```
