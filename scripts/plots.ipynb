```python
def exponential(x, a, b):
    return a * np.exp(-b * x)

def power(x, a, b):
    return a * (x ** b)
```

```python
# MEASURE QUALITY OF FIT USING R^2 AND CHI^2 METRIC
```

```python
def r_2(y, y_fit):
    ss_res = np.sum((y - y_fit) ** 2)
    ss_tot = np.sum((y - np.mean(y)) ** 2)
    return 1 - (ss_res / ss_tot)
```

```python
log_anti_r2 = r_2(averaged_number_of_unique_anti_10, log(x_anti, *log_anti_params))
power_anti_r2 = r_2(averaged_number_of_unique_anti_10, power(x_anti, *power_anti_params))

print(f"R^2 for anti hubs for log fit: {log_anti_r2}")
print(f'R^2 for anti hubs for power fit: {power_anti_r2}\n')
```

```python
log_pro_r2 = r_2(averaged_number_of_unique_pro_10, log(x_pro, *log_pro_params))
power_pro_r2 = r_2(averaged_number_of_unique_pro_10, power(x_pro, *power_pro_params))

print(f"R^2 for pro hubs for log fit: {log_pro_r2}")
print(f'R^2 for pro hubs for power fit: {power_pro_r2}\n')
```
